```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('Universities.csv')
```

```python
df1
```

```python
plt.hist(df1['SAT'])
plt.xlabel('SAT')
plt.ylabel('Frequency')
plt.title('Distribution of SAT Scores')
plt.show()
```

```python
sns.histplot(df1['SAT'], color='brown')
plt.xlabel('SAT')
plt.ylabel('Frequency')
plt.title('Distribution of SAT Scores (Brown)')
plt.show()
```

```python
print("Mean of SAT scores:", df1['SAT'].mean())
print("Median of SAT scores:", df1['SAT'].median())
print("Mode of SAT scores:", df1['SAT'].mode())
```

```python
plt.plot(df1['SAT'], df1['Accept'],'rs')
plt.xlabel('Top 10')
plt.ylabel('Accept')
plt.title('SAT vs. Acceptance Rate')
plt.show()
```

```python
sns.pairplot(df1)
```

```python
sns.pairplot(df1.iloc[:,1:3])
```

```python
plt.boxplot(df1['SAT'])
plt.show()
```
